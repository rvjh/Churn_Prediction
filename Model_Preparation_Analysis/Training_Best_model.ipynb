```python
import pandas as pd
import numpy as np
```

```python
df_clean = pd.read_csv("../Clean_Input_Data/credit_card_churn_clean.csv")
df_clean
```

```python
df_clean.columns
```

```python
df_clean.describe()
```

```python
df_clean.dtypes
```

```python
print(df_clean.gender.unique())
print(df_clean.marital_status.unique())
print(df_clean.card_category.unique())
print(df_clean.active_customer.unique())
print(df_clean.active_customer.unique().sum())

df_clean['gender'] = df_clean.gender.replace({
    'M':0,
    'F':1
})

df_clean['marital_status'] = df_clean.marital_status.replace({
    'married':0, 
    'single':1, 
    'divorced':2
})

df_clean['card_category'] = df_clean.card_category.replace({
    'blue':0, 
    'gold':1, 
    'silver':2, 
    'platinum':3
})

df_clean['active_customer'] = df_clean.active_customer.replace({
    True:0,
    False:1
})
```

```python
df_clean
```

```python
X= df_clean.drop(columns=["Unnamed: 0","active_customer","customer_id"])
y=df_clean['active_customer']

X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X= scaler.fit_transform(X)

X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=20)
```

```python
print(f"Shape of X_train is : {X_train.shape}")
print(f"Shape of X_test is : {X_test.shape}")
print(f"Shape of y_train is : {y_train.shape}")
print(f"Shape of y_test is : {y_test.shape}")
```

```python
import xgboost
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
algos = [LogisticRegression(), 
         DecisionTreeClassifier(), 
         RandomForestClassifier(), 
         XGBClassifier()]
names = ["Logistic Regression",
         "DecisionTree Classifier",
         "RandomForest Classifier", 
         "XGBoost Classifier"]
acc = []
```

```python
for name in algos:
    model = name
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    accu = accuracy_score(y_test,y_pred)
    acc.append(accu) 
```

```python
acc
```

```python
evaluation = pd.DataFrame({'Name':names, 'Accuracy_Score':acc})
evaluation
```

## Hyparameter Tuning using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
xgb_model= XGBClassifier()
```

```python
params = {
    'n_estimators': [100,200,300],
    'max_depth': [1,2,3,4],
    'eta': [0.1,0.01,0.001]
}
```

```python
cross_val = GridSearchCV(estimator=xgb_model,param_grid=params,verbose=2)
cross_val.fit(X_train,y_train)
```

```python
cross_val.best_estimator_
```
